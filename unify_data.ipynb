```python
import pandas as pd
import os
import re
```

```python
#Para información sobre en qué unidades está cada cosa estan los archivos
datos = pd.DataFrame()
datos["fecha"] = []
datos["hora"] = []
datos["estacion"] = []
datos["temperatura"] = []
datos["humedad"] = []
datos["presion"] = []
datos["direccion_viento"] = []
datos["velocidad_viento"] = []
datos.head()
```

```python
directorio_datos = os.scandir("datos_descargados")
for archivo in directorio_datos:
    try:
        b = 2
        with open(f"datos_descargados/{archivo.name}",mode="r",encoding="cp1252") as f:
            for line in f.readlines():
                if b > 0:
                    b = b-1
                    continue
                s = re.split("\s+",line,maxsplit=7)
                if len(s) < 8:
                    continue
                fecha = s[0]
                hora = s[1]
                temp = s[2]
                hum = s[3]
                pre = s[4]
                direc = s[5]
                vel = s[6]
                estacion = s[7]
                datos = datos.append({"fecha":fecha,"hora":hora,"estacion":estacion,	"temperatura":temp,"humedad":hum,"presion":pre,"direccion_viento":direc,"velocidad_viento":vel},ignore_index=True)
    except:
        print(archivo.name)
        break


datos.head()
```
